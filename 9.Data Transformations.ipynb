## Scaling Discussion

**Standard Scaling** is preferred when the data follows a normal distribution or when the algorithm is sensitive to outliers (e.g., Logistic Regression, SVM). It transforms data to have mean=0 and std=1.

**MinMax Scaling** is preferred when you need values in a fixed range [0,1], especially for algorithms like Neural Networks or KNN. However, it is sensitive to outliers.

## Encoding Discussion

**One-Hot Encoding (OHE)**
- Pros: No ordinal relationship assumed, works well for low cardinality columns
- Cons: Creates many new columns for high cardinality columns, increases dimensionality

**Label Encoding (LE)**
- Pros: Simple, keeps data compact, works well for high cardinality columns
- Cons: Assigns numeric order to categories which may mislead the model

## Feature Engineering Discussion

**capital_net**: Difference between capital_gain and capital_loss. 
Captures the net financial activity of a person which is more meaningful than two separate columns.

**age_group**: Binning age into Young, Middle, Senior groups helps the model 
capture non-linear relationships between age and income.

**Log Transformation on capital_gain**: capital_gain is heavily right-skewed 
with most values at 0 and a few extreme values up to 99999. 
log1p transformation compresses the scale and reduces skewness, 
making it easier for the model to learn from this feature.
